```python
from keras.layers import *
from keras.models import Model

#define two sets of input
inputA = Input(shape = (32,))
inputB = Input(shape=(128,))

# the first branch operations on the first input
x= Dense (8, activation="relu")(inputA)
x1= Dense (4, activation="relu")(x)

# the second branch opetates on the second input
y= Dense(64, activation="relu")(inputB)
y1 = Dense(32, activation="relu")(y)
y2 = Dense(4, activation="relu")(y1)

#Combine outputs of the two branches
combined = concatenate ((x1,y2))

#apply a FC lauyer and then a regression prediction on the combine outputs
z= Dense (2,activation = "relu")(combined)
z1 = Dense(1,activation = "linear")(z)

# Our model will accept the inputs of the two branches and then output a single value
model = Model(inputs = (inputA, inputB), outputs = z1)
model.summary()
```
