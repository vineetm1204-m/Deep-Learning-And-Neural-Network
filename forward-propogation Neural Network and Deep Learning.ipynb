```python
import numpy as np
```

```python
i1, i2 = 0.05, 0.10
t1, t2 = 0.01, 0.99

w1, w2, w3, w4 = 0.15, 0.20, 0.25, 0.30
w5, w6, w7, w8 = 0.40, 0.45, 0.50, 0.55

b1, b2 = 0.35, 0.60
lr = 0.5

```

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))

h1 = sigmoid(i1*w1 + i2*w3 + b1)
h2 = sigmoid(i1*w2 + i2*w4 + b1)

o1 = sigmoid(h1*w5 + h2*w7 + b2)
o2 = sigmoid(h1*w6 + h2*w8 + b2)

E = 0.5*(t1 - o1)**2 + 0.5*(t2 - o2)**2

d_o1 = (o1 - t1) * o1 * (1 - o1)
d_o2 = (o2 - t2) * o2 * (1 - o2)

w5 -= lr * d_o1 * h1
w6 -= lr * d_o2 * h1
w7 -= lr * d_o1 * h2
w8 -= lr * d_o2 * h2

d_h1 = (d_o1*w5 + d_o2*w6) * h1 * (1 - h1)
d_h2 = (d_o1*w7 + d_o2*w8) * h2 * (1 - h2)

w1 -= lr * d_h1 * i1
w2 -= lr * d_h2 * i1
w3 -= lr * d_h1 * i2
w4 -= lr * d_h2 * i2

print("Total Error:", E)
print("Updated Weights:")
print(w1, w2, w3, w4, w5, w6, w7, w8)
```
